```python
# ============================================
# LAYER 2 - CELL 18: LSTM SEQUENCE PREDICTOR
# Deep learning for time-series patterns
# ============================================

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import MinMaxScaler

# Prepare sequence data for LSTM
def create_sequences(data, lookback=20):
    """Create sequences for LSTM training"""
    X, y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i-lookback:i])
        y.append(data[i])
    return np.array(X), np.array(y)

# Use BTC data (largest dataset)
btc_data = df_combined[df_combined['symbol'] == 'BTC'][feature_cols + ['target']].copy()

# Scale features
scaler = MinMaxScaler()
btc_scaled = scaler.fit_transform(btc_data[feature_cols])

# Create sequences
X_seq, y_seq = create_sequences(btc_scaled, lookback=20)
y_labels = btc_data['target'].values[20:]  # Target labels

# Train/test split
split = int(len(X_seq) * 0.8)
X_train_lstm = X_seq[:split]
X_test_lstm = X_seq[split:]
y_train_lstm = y_labels[:split]
y_test_lstm = y_labels[split:]

# Build LSTM model
model_lstm = keras.Sequential([
    layers.LSTM(64, return_sequences=True, input_shape=(20, len(feature_cols))),
    layers.Dropout(0.2),
    layers.LSTM(32),
    layers.Dropout(0.2),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train
print(" Training LSTM...")
history = model_lstm.fit(
    X_train_lstm, y_train_lstm,
    epochs=20,
    batch_size=16,
    validation_split=0.2,
    verbose=0
)

# Evaluate
y_pred_lstm = (model_lstm.predict(X_test_lstm, verbose=0) > 0.5).astype(int).flatten()
accuracy_lstm = accuracy_score(y_test_lstm, y_pred_lstm)

print(f"\n📊 LSTM Performance:")
print(f"   Training: {len(X_train_lstm)} sequences")
print(f"   Test: {len(X_test_lstm)} sequences")
print(f"   Accuracy: {accuracy_lstm*100:.1f}%")
print(f"\n🏆 Model Comparison:")
print(f"   XGBoost: {accuracy_multi*100:.1f}%")
print(f"   LSTM: {accuracy_lstm*100:.1f}%")
```
